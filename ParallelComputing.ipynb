```python
import threading
import time

def proses(angka):
    time.sleep(1)
    print(f"Data {angka} -> {angka * 3}")

print("=== PARALLEL COMPUTING ===")
data = [1, 2, 3, 4]

start = time.time()

# Membuat thread untuk setiap data
threads = []
for i in data:
    t = threading.Thread(target=proses, args=(i,))
    threads.append(t)
    t.start()

# Menunggu semua thread selesai
for t in threads:
    t.join()

print(f"Waktu: {time.time()-start:.2f} detik")
```

```python
import time

# Serial Computing - Proses berurutan
def serial_computing():
    print("=== SERIAL COMPUTING ===")
    print("Proses dijalankan satu per satu secara berurutan")
    
    data = [1, 2, 3, 4]
    
    start = time.time()
    for angka in data:
        time.sleep(1)  # Simulasi proses 1 detik
        hasil = angka * 2
        print(f"Data {angka} -> Hasil: {hasil}")
    
    print(f"Total waktu: {time.time()-start:.2f} detik")

serial_computing()
```

```python
# SISD - Satu instruksi, satu data
def sisd_sederhana():
    print("=== SISD ===")
    print("Satu instruksi dijalankan untuk satu data")
    
    data = 5
    instruksi = data * 2  # Satu instruksi perkalian
    
    print(f"Data: {data}")
    print(f"Instruksi (kali 2): {instruksi}")
    
sisd_sederhana()
```

```python
# SIMD - Satu instruksi untuk banyak data sekaligus
def hitung_kuadrat_simd(angka):
    print("=== SIMD (Single Instruction Multiple Data) ===")
    print("Satu instruksi (kuadrat) untuk BANYAK data")
    
    # Simulasi SIMD: instruksi yang sama untuk semua data
    hasil = []
    for a in angka:
        hasil.append(a ** 2)  # Instruksi kuadrat untuk setiap data
        print(f"Data {a} -> {a**2}")
    
    return hasil

angka = [2, 3, 4, 5]
hitung_kuadrat_simd(angka)
```

```python
# MISD - Banyak instruksi berbeda untuk satu data
def misd_sederhana(data):
    print("=== MISD (Multiple Instruction Single Data) ===")
    print("Banyak instruksi berbeda untuk SATU data yang sama")
    
    angka = data
    print(f"Data tunggal: {angka}")
    
    # Instruksi 1: Kuadrat
    kuadrat = angka ** 2
    print(f"Instruksi 1 (kuadrat): {kuadrat}")
    
    # Instruksi 2: Akar kuadrat
    akar = angka ** 0.5
    print(f"Instruksi 2 (akar): {akar:.2f}")
    
    # Instruksi 3: Kali 2
    kali_dua = angka * 2
    print(f"Instruksi 3 (kali 2): {kali_dua}")

misd_sederhana(8)
```

```python
import threading

def genap(angka):
    print(f"Genap: {angka} kuadrat = {angka**2}")

def ganjil(angka):
    print(f"Ganjil: {angka} kubik = {angka**3}")

print("=== MIMD ===")

# Instruksi berbeda untuk data berbeda
t1 = threading.Thread(target=genap, args=(2,))
t2 = threading.Thread(target=ganjil, args=(3,))

t1.start()
t2.start()
t1.join()
t2.join()
```
